## 5 To avoid all kind of type issues, we like to (a) convert all numerical values to float(prices, quantities …etc), and (b) all string values should be lowercased.

```python
def type_converter(col):
    if col.dtype == 'object':
        col = col.astype('str').str.lower()
    else:
        col = col.astype('float64')
    return col
# data.apply(lambda x : x.astype('str').str.lower() if x.dtype == 'object' else x.astype('float64'))
data=data.apply(type_converter,axis=0)
```

## 6 We like to codify sales into new column named “sale interest” which consists ofthree values (1: basic, 2: moderate, 3: high).

```python
data.Sales.describe()
boundary_25 = data.Sales.describe()['25%']
boundary_75 = data.Sales.describe()['75%']
data["sale interest"] = np.nan
def codify_sales(row):
    
    if row.Sales > boundary_75 :
        row['sale interest'] = "high"
    elif row.Sales > boundary_25:
        row['sale interest'] = "moderate"
    else:
        row['sale interest'] = "low"
    
    return row
data=data.apply(codify_sales,axis='columns')
```

## 7 We like to normalize sale by dividing on the highest sale value (keep values in 0 to 1). 

```python
highest_sale = data.Sales.max()
data.Sales = data.Sales.map(lambda x : x / highest_sale)
data
```

```python
#adding new featute to measure the ship mode
def tf_ship_mode(mode):
    if mode == 'standard class':
        return 1
    elif mode =='second class':
        return 2
    elif mode == 'first class':
        return 3
    else : 
        return 0
    
modified_data = data
modified_data['Ship Weights'] = modified_data['Ship Mode'].map(lambda mode : tf_ship_mode(mode))
modified_data.head()
```

```python
grp_cust_prod = modified_data.groupby(['Customer ID','Product ID'])
#calculate mean sales
mean_grp_sales=grp_cust_prod.Sales.mean()
print(mean_grp_sales)
```

```python
#calculate mean ship
mean_grp_ship=grp_cust_prod['Ship Weights'].mean()
print(mean_grp_ship.describe())
```

```python
#fucntion to construct {shipping_mode:val , sales:val } from a given customer and a given product
def gen_stat(customer_id,product_id):
    mean_sale=mean_grp_sales.loc[customer_id].loc[product_id]
    return {'mean_sale':mean_sale}
```
