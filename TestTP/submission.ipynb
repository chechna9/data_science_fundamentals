```python
import pandas as pd
from sklearn.model_selection import train_test_split
```

1-Loading the dataset and ommiting the timestamp

```python
data = pd.read_csv('./ratings.csv')
data.drop(columns='timestamp',inplace=True)
```

```python
data
```

2- splitting data to train test split

```python
train,test = train_test_split(data,random_state=1,train_size=0.8,test_size=0.2)
```

```python
train
```

```python
test
```

3- Converting train and test to dense tables

```python
training_data = train.pivot(index='userId',columns=['movieId'],values='rating')
training_data
```

```python
test_data = test.pivot(index='userId',columns=['movieId'],values='rating')
test_data
```

4 - calulating the golbal mean of ratings

```python
global_mean = data.rating.mean()
global_mean
```

5 - users's bias

```python
def calculate_bu(row:pd.Series,lamp):
    lu = row.loc[row.notna()]
    len_lu = len(lu)
    return  (lu.sum() - len_lu * global_mean ) / (lamp + len_lu)
    
```

```python
# testing it
calculate_bu(training_data.iloc[15],lamp=0.99)
```
